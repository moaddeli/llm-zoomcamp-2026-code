```python
from gitsource import GithubRepositoryDataReader
```

```python
reader = GithubRepositoryDataReader(
    repo_owner="DataTalksClub",
    repo_name="llm-zoomcamp",
    commit_id="8c1834d",
    allowed_extensions={"md"},
    filename_filter=lambda path: "/lessons/" in path,
)

files = reader.read()
```

```python
print(files[0])
```

```python
documents = []

for file in files:
    doc = file.parse()
    documents.append(doc)
```

```python
# Q1. How many lesson pages
print(f"Number of lesson pages= {len(documents)}")
```

```python
from minsearch import Index

index = Index(
    text_fields=["content"], 
    keyword_fields=["filename"]
)

index.fit(documents)
```

```python
question ="How does the agentic loop keep calling the model until it stops?"
search_results = index.search(
    question,
    num_results=25
)
```

```python
for search_result in search_results:
    print(search_result["filename"])
```

## Q6. Turning it into an agent


```python
from toyaikit.llm import OpenAIClient, OpenAIChatCompletionsClient
from toyaikit.tools import Tools
from toyaikit.chat import IPythonChatInterface
from toyaikit.chat.runners import OpenAIResponsesRunner, DisplayingRunnerCallback, OpenAIChatCompletionsRunner
```

```python
def search(query: str) -> dict[str, str]:
    """
    Search the content of chuncks of file for entries matching the given query.
    """
    boost_dict = {"content": 1.0}


    return index.search(
        query,
        num_results=7,
        boost_dict=boost_dict,
    )
```

```python
agent_tools = Tools()
agent_tools.add_tool(search)
```

```python
agent_tools.get_tools()
```

```python
instructions = """
You're a course teaching assistant. Answer the student's question using the search tool. 
Make multiple searches with different keywords before answering.
""".strip()
```

```python
chat_interface = IPythonChatInterface()
callback = DisplayingRunnerCallback(chat_interface)

runner = OpenAIChatCompletionsRunner(
    tools=agent_tools,
    developer_prompt=instructions,
    chat_interface=chat_interface,
    llm_client=OpenAIChatCompletionsClient(model=MODEL_NAME, client=openai_client)
)
```

```python
result = runner.loop(
    prompt="How does the agentic loop work, and how is it different from plain RAG?",
    callback=callback,
)
```
